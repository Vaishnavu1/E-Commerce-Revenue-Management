## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5

```python
ECom_Data.head()
```

```python
#Total sum of discounted price greater than 600 
a= pd.DataFrame(ECom_Data[ECom_Data['discounted_price']>600])
df = a['discounted_price']*0.25
df.sum()
```

```python
#Total sum of discounted price greater than 350 and less than or equal to 600
b = pd.DataFrame(ECom_Data[(ECom_Data['discounted_price']>350) & (ECom_Data['discounted_price']<=600)])
df1 = b['discounted_price']*0.15
df1.sum()
```

```python
#Total sum of discounted price greater than 100 and less than or equal to 350
c = pd.DataFrame(ECom_Data[(ECom_Data['discounted_price']>100) & (ECom_Data['discounted_price']<=350)])
df2 =  c['discounted_price']*0.10
df2.sum()
```

```python
#Total sum of discounted price less than or equal to 100
d = pd.DataFrame(ECom_Data[ECom_Data['discounted_price']<=100])
df3 =  d['discounted_price']*0.05
df3.sum()
```

```python
net_revenue = df.sum()+df1.sum()+df2.sum()+df3.sum()
```

```python
print('The total net revenue generated by the E-Commerce company over all orders placed is',net_revenue)
```

## Q6.Calculate the Net BrandRevenue for each Brand and list the top 14 Brand having maximum revenue in descending order (6 marks)
#### Hint: Net BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's net revenue.

### Ans 6

```python
#revenue of brands where discounted price is greater than 600
rev1= a['discounted_price']-df
#Assigning a new cloumn to dataset a (created for price greater than 600)
a['net_brand_revenue']=rev1
#Grouping net brand revenue based on brand where discounted price is greater than 600
Revenue1 =a.groupby('Brand')['net_brand_revenue'].sum()
```

```python
#revenue of brands where discounted price is greater than 350 and less than or equal to 600 
rev2 = b['discounted_price']-df1
#Assigning a new cloumn to dataset b(created for price greater than 350 and less than or equal to 600)
b['net_brand_revenue']=rev2
#Grouping net brand revenue based on brand where discounted price is greater than 350 and less than or equal to600
Revenue2= b.groupby('Brand')['net_brand_revenue'].sum()
```

```python
#revenue of brands where discounted price is greater than 100 and less than or equal to 350 
rev3 = c['discounted_price'] - df2
#Assigning a new cloumn to dataset b(created for price greater than 100 and less than or equal to 350 )
c['net_brand_revenue']=rev3
#Grouping net brand revenue based on brand where discounted price is greater than 100 and less than or equal to 350
Revenue3 = c.groupby('Brand')['net_brand_revenue'].sum()
```

```python
#Merging 3 different dataframes to one to obtain total revenue as per brands
Revenue001=pd.DataFrame(pd.merge(Revenue1,Revenue2,on='Brand',how='outer'))
```

```python
#Merging 3 different dataframes to one to obtain total revenue as per brands
Revenue =pd.DataFrame(pd.merge(Revenue001,Revenue3,on='Brand',how='outer')).fillna(0)
```

```python
Revenue
```

```python
#Adding a new column(Total net brand revenue) to the obtained data
Revenue['Total_Brand_Revenue']=Revenue['net_brand_revenue_x']+Revenue['net_brand_revenue_y']+Revenue['net_brand_revenue']
```

```python
# Net BrandRevenue for each Brand
Revenue.reset_index()
```

```python
# Top 14 brands having more revenue :
Revenue.sort_values(by='Total_Brand_Revenue',ascending=False).head(14)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly net Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days

### Ans 9 a)

```python

# For providing mothly net revenue of E-Commerce Company we need to create a dataset showing order placed monthwise and the revenue generated by company in each order.
# a ,b &c are the datasets were we have arranged the data based on company net revenue.
# converting the date column to datetime format
a['Order_Date'] = pd.to_datetime(a['Order_Date'],dayfirst=True)
b['Order_Date'] = pd.to_datetime(b['Order_Date'],dayfirst=True)
c['Order_Date'] = pd.to_datetime(c['Order_Date'],dayfirst=True)
```

```python
#converted order date 
a.head()
```

```python
b.head()
```

```python
c.head()
```

```python
# Adding a new month column (based on order date) in a,b & c dataset
a['Month']= a['Order_Date'].dt.month
b['Month']= b['Order_Date'].dt.month
c['Month']= c['Order_Date'].dt.month
```

```python
# Adding a new column of revenue earned by company based on Q5 in each data set
a['Net_Company_revenue']=df
b['Net_Company_revenue']=df1
c['Net_Company_revenue']=df2
```

```python
a.head()
```

```python
# Now the data in a,b&c are concatinated to form a new data set 
New_Data= pd.concat([a,b,c],keys=['Data_A','Data_B','Data_C'])
```

```python
New_Data.head()
```

```python
# The month column is now corrected from month number to minth name
New_Data['Month'] = pd.to_datetime(New_Data['Month'], format='%m').dt.month_name().str.slice(stop=3)
```

```python
New_Data.head()
```

```python
New_Data.sort_values(by='Order_Date',inplace=True)
```

```python
New_Data
```

```python
#Creating Line plot with the new data set for each region with net company revenue
plt.figure(figsize=(20,7))
sns.lineplot(data=New_Data,x='Month',y='Net_Company_revenue',ci = False,hue='Region')
plt.title('LINE PLOT-MONTLY NET_REVENUE OF E-COMMERCE COMPANY FOR EACH REGION')
plt.legend(bbox_to_anchor=[1, 1]);
```
